## Use Newton's Method for Optimization 

```
import numpy as np

# import data
x = np.array(np.loadtxt('q1x.dat'))
y = np.array(np.loadtxt('q1y.dat'))
y = np.reshape(y,(len(y),1))
x0 = [1]*len(x)
x0 = np.array(x0)
x0 = np.reshape(x0, (len(x0),1))
x = np.append(x0,x, axis = 1) # add a constant 1 for x0 in x
```

```
#initialize parameter theta to be a 0 vectror
theta=np.zeros((3,1))

for step in range(0,20000):
  h = 1/(1 + np.exp(-theta[0,0]*x[:,0]-theta[1,0]*x[:,1]-theta[2,0]*x[:,2]))
  h = np.reshape(h,(len(h),1))
 
  lp = np.sum((h-y)*x, axis = 0) #first derivatives
  if lp[0]>-0.1 and lp[1]>-0.1 and lp[2]>-0.1:
    break

  H = np.zeros((3,3))
  for i in range(0,3):
    for j in range(0,3):
      H[i,j] = np.sum(h*(1-h)*x[:,i]*x[:,j])

  H_inv = np.linalg.inv(H) 
  
  theta = theta - np.reshape(np.dot(H_inv,lp),(3,1))
```
